Sinais de EEG de 9 usuários

Sinais de entrada consistem em 4 tipos sendo eles: mão esquerda (classe 1), mão direita (classe 2), dois pés (classe 3) e língua (classe 4).

2 sessões em diferentes dias foram gravadas para cada usuários. Cada sessão é composta por 6 coletas separadas por uma pequena pausa. Cada coleta corresponde à 48 trials (12 de cada classe) = 288 trials por sessão.

No início de cada sessão ocorria uma gravação de aproximadamente 5 minutos para estimação da influẽncia do EOG (). A gravação era dividida em 3 blocos:

* 1° - 2 minutos com olhos abertos olhando para uma cruz fixa na tela
* 2° - 1 minuto de olhos fechados
* 3° - 1 minuto movimentando os olhos

Para o usuário A04T ocorreu problemas técnicos com o bloco de EOG, e por isso, possui apenas o bloco do movimento dos olhos.

Usuários ficam em uma poltrona em frente a uma tela de computador. Em t=0 uma cruz fixa aparece na tela com fundo preto e toca um som. Após 2 segundos aparece um sinal de entrada na forma de uma flecha apontando para alguma direção (esq, dir, cima e baixo) que permanece por 1,25s. Os usuários devem imaginar a tarefa motora após o sinal durante 6s.

22 eletrodos Ag/AgCl (com distância de 3,5cm entre eles) foram utilizados. Usaram o mastoide esquerdo como referência e o direito como ground.

Sinais foram amostrados em 250 Hz e utilizaram um filtro passa faixas de 0,5Hz a 100 Hz. Sensibilidade do amplificador foi configurada para 100uV. Filtro notch de 50 Hz.

Montagem seguindo o sistema internacional 10-20.

Além dos 22 eletrodos para EEG utilizaram 3 monopolares canais de EOG amostrados em 250Hz, filtrados em 0,5 Hz a 100 Hz, com filtro notch em 50Hz e sensibilidade de 1mV.


#### Descrição dos dados dos arquivos

Dados armazenados em GDF (General Data Format) para sinais biomédicos, um arquivo para usuário e sessão. Apenas uma sessão contém os rótulos das classes para todos os trials.

Dados organizados em [s, h] = sinais e cabeçalho. Sinais contém 25 cnanais (22 EEG e 3 EOG, sendo estes os 3 últimos). Para o cabeçalho temos 3 campos, sendo eles, de posição (h.EVENT.POS), o tipo (h.EVENT.TYP) e de duração (h.EVENT.DUR). Classes 1, 2, 3 e 4 correspondem, respectivamente, aos eventos 769, 770, 771 e 772. Os trials que possuem artefatos como pontuações de experts estão marcadas como eventos do tipo 1023. h.ArtifactSelection contém uma lista de todas os trials, onde 0 a um trial limpo e 1 a um trial com artefato.


#### Avaliação

Participantes devem apresentar uma resposta para cada amostra, sendo ela com artefato ou não. Uma matriz de confusão será construída para cada trial sem artefato, e por meio da matriz, será obtido o coeficiente de Kappa e a acurácia.

### Preprocessing

```python
from braindecode.preprocessing import preprocess, Preprocessor
import numpy as np
from numpy import multiply

# Parameters for filtering
low_cut_hz = 4.  # low cut frequency for filtering
high_cut_hz = 38.  # high cut frequency for filtering

# Paramater for resampling
#newfreq = 128

# Parameter for scaling
factor = 1e6

preprocessors = [
    Preprocessor('pick_types', eeg=True, meg=False, stim=False),  # Keep EEG sensors
    Preprocessor(lambda data: multiply(data, factor)),  # Convert from V to uV
    # Preprocessor("resample", sfreq=newfreq), # Resampling - if commented it will not perform resampling
    Preprocessor('filter', l_freq=low_cut_hz, h_freq=high_cut_hz),  # Bandpass filter
    Preprocessor("set_eeg_reference", ref_channels="average", ch_type="eeg") # Common Average Reference
]

# Transform the data
preprocess(dataset, preprocessors, n_jobs=-1);
```

## Create model

```python
valid_set
```

```python
import torch
from braindecode.util import set_random_seeds
from braindecode.models import EEGNetv4

cuda = torch.cuda.is_available()  # check if GPU is available, if True chooses to use it
device = 'cuda' if cuda else 'cpu'
if cuda:
    torch.backends.cudnn.benchmark = False

seed = 20200220
np.random.seed(seed)
set_random_seeds(seed=seed, cuda=cuda)

n_classes = np.unique(train_set.datasets[0].y).size
# Extract number of chans and time steps from dataset
n_chans = train_set[0][0].shape[0]
input_window_samples = train_set[0][0].shape[1]
input_window_seconds = input_window_samples / sfreq

F1, D = 8, 2

# Setting the length of the temporal kernel at half the sampling rate allows for
# capturing frequency information at 2Hz and above.
kernel_length = 64 # sfreq//4 = 62

pool1size = 8
dropout_prob = 0.5

model = EEGNetv4(
    n_chans,
    n_classes,
    n_times=input_window_samples,
    final_conv_length='auto',
    F1=8,
    D=2,
    F2=F1*D,
    kernel_length=kernel_length,
    pool1_kernel_size=pool1size,
    drop_prob=dropout_prob,
)
model.to(device);
```

## Training

```python
def combine_datasets(dataset):
    """
    Combines data and targets from all WindowsDatasets within a given dataset.

    Args:
        dataset (BaseConcatDataset): The dataset to combine.

    Returns:
        tuple: A tuple containing the concatenated data and targets.
    """

    all_X, all_y = [], []
    for ds in dataset.datasets:
        for data, y, _ in ds:
            all_X.append(data)
            all_y.append(y)

    return np.stack(all_X, axis=0), np.stack(all_y, axis=0)
```

```python
X_train, y_train = combine_datasets(train_set)
```

```python
other_X_train, other_y_train = combine_datasets(other_train_set)
```

```python
other_X_train.shape, other_y_train.shape
```

```python
X_train.shape, y_train.shape
```

```python
from skorch.callbacks import LRScheduler
from skorch.helper import predefined_split

from braindecode import EEGClassifier

batch_size = 64
n_epochs = 100

clf = EEGClassifier(
    model,
    criterion=torch.nn.CrossEntropyLoss,
    optimizer=torch.optim.Adam,
    train_split=predefined_split(valid_set),  # using valid_set for validation
    # optimizer__lr=lr,
    # optimizer__weight_decay=weight_decay,
    batch_size=batch_size,
    callbacks=[
        "accuracy", ("lr_scheduler", LRScheduler('CosineAnnealingLR', T_max=n_epochs - 1)),
    ],
    device=device,
)
# Model training for a specified number of epochs. `y` is None as it is already supplied
# in the dataset.
clf.fit(X_train, y=y_train, epochs=n_epochs);
```

\We can then compute the accuracy on the **evaluation (test) set**:


```python
X_test, y_test = combine_datasets(test_set)
```

```python
import numpy as np
print(f"Mean Train Accuracy: {np.mean(clf.predict(X_train) == [y for y in y_train])*100:.2f}%")
```

```python
X_val, y_val = combine_datasets(valid_set)
print(f"Mean Validation Accuracy: {np.mean(clf.predict(X_val) == [y for y in y_val])*100:.2f}%")
```

```python
print(f"Mean Test Accuracy: {np.mean(clf.predict(X_test) == [y for y in y_test])*100:.2f}%")
```

```python
annotations = ['feet', 'left_hand', 'right_hand', 'tongue']
preds = clf.predict(test_set.datasets[0])
accuracies = {}
for event, y_event, _ in test_set.datasets[0]:
  trial_mask = np.array([y == y_event for X,y,i in test_set.datasets[0]])
  acc = np.mean(preds[trial_mask] == y_event)
  if y_event not in accuracies:
    accuracies.setdefault(y_event, [])
  accuracies.get(y_event).append(acc)
```

```python
for i in range(len(annotations)):
  print(f"{annotations[i]}: {np.mean(accuracies.get(i))*100:.2f}%")
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix

conf_matrix = confusion_matrix(
    y_true=[annotations[y] for _,y,_ in test_set.datasets[0]],
    y_pred=[annotations[i] for i in preds],
    normalize='true'
    )

disp = ConfusionMatrixDisplay(
    confusion_matrix=conf_matrix,
    display_labels=annotations
)

disp.plot()
plt.show();
```
